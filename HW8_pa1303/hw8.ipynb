```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import os
%matplotlib inline
```

```python
df = pd.read_csv(os.getenv('PUIDATA')+'/2012_SAT_Results.csv')
```

```python
df.head()
```

```python
df['SAT Critical Reading Avg. Score'] = pd.to_numeric(df['SAT Critical Reading Avg. Score'], errors='coerce')
```

```python
df['SAT Math Avg. Score'] = pd.to_numeric(df['SAT Math Avg. Score'], errors='coerce')
```

```python
df['Num of SAT Test Takers'] = pd.to_numeric(df['Num of SAT Test Takers'], errors='coerce')
```

```python
df = df.dropna()
```

```python
x = df['SAT Critical Reading Avg. Score']
y = df['SAT Math Avg. Score']
area = df['Num of SAT Test Takers']
```

```python
ax = plt.figure(figsize=(16,12))
plt.scatter(x, y, s = area)
plt.xlabel('Critical Reading Score')
plt.ylabel('Math Score')
plt.title('School performance on SAT')
```

#### The SAT scores of Reading vs Math for each school in NYC. The size of the dots represent the number of students in the school
